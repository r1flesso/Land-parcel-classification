Разделим классы на редкие и нередкие: классы 2, 6, 1 будут нередкими, а классы 4, 3, 5, собственно, редкими

```python
df_B = df[~df['id_2'].isin([4, 3, 5])]
df_B.to_csv('dataset_B.csv', index=False)
```

```python
df_C = df[~df['id_2'].isin([1, 2, 6])]
df_C.to_csv('dataset_C.csv', index=False)
```

```python
df['rarity'] = df['id_2'].apply(lambda x: 0 if x in [2, 6, 1] else 1)
df = df.drop(columns=['id_2'])
df.to_csv('dataset_A.csv', index=False)
df.head()
```
