```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
os.chdir('..')
df = pd.read_csv('src/clear_data.csv')
df = df.dropna()
```

```python
df['name'].unique()
```

```python
color_map = {
    'орошение': '#0d47a1',
    'лес': "#246d26",
    'НП': '#424242',
    'богарные': '#5d4037',
    'сенокосы': '#37474f',
    'не сх': '#b71c1c'
}

plt.figure(figsize=(22, 12))

sns.scatterplot(
    x='X',
    y='Y',
    hue='name',
    data=df,
    palette=color_map,
    s=3
)

legend = plt.legend(fontsize=15)

for handle in legend.legend_handles:
    handle.set_markersize(10)

plt.title('Территории Марксовского района', fontweight='bold', fontsize=30)
plt.xlabel('Координата X', fontsize=20)
plt.ylabel('Координата Y', fontsize=20)

plt.show()
```

```python
boxes = [col for col in df.columns if df[col].dtype == float and col not in 'XY']
print(len(boxes), boxes)
```

```python
fig, axes = plt.subplots(6, 6, figsize=(20, 15))
axes_flat = axes.flatten()

for i, col in enumerate(boxes):
    if i < 36:
        axes_flat[i].boxplot(df[col].dropna())
        axes_flat[i].set_title(col)

plt.tight_layout()
plt.show()
```
