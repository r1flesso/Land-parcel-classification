```python
import pandas as pd
import os
```

```python
os.chdir('..')
df = pd.read_csv('scripts/hyperparameters_n_results.csv')
df.head()
```

```python
df_ranks = pd.DataFrame({
    'run_name': df['run_name'],
    'f1_rank': df['f1_score'].rank(ascending=False).astype(int),
    'precision_rank': df['precision'].rank(ascending=False).astype(int),
    'recall_rank': df['recall'].rank(ascending=False).astype(int),
    'kappa_rank': df['cohen_kappa'].rank(ascending=False).astype(int),
    'mcc_rank': df['matthews_corrcoef'].rank(ascending=False).astype(int),
    'roc_auc_rank': df['roc_auc_score'].rank(ascending=False).astype(int)
})

df_ranks.head(12)
```

Тут мы видим, что самые лучшие модели это:
1. RFC_with_sqrt_n_False_bootstrap_n_roc_auc_score (естественно)
2. RFC_with_sqrt_n_True_bootstrap_n_cohen_kappa
3. RFC_with_sqrt_n_False_bootstrap_n_matthews_corrcoef
4. RFC_with_sqrt_n_False_bootstrap_n_precision

```python
best_models = [
    'RFC_with_sqrt_n_False_bootstrap_n_roc_auc_score',
    'RFC_with_sqrt_n_True_bootstrap_n_cohen_kappa',
    'RFC_with_sqrt_n_False_bootstrap_n_matthews_corrcoef',
    'RFC_with_sqrt_n_False_bootstrap_n_precision'
]

df[df['run_name'].isin(best_models)].head()
```

Тут представлены гиперпараметры 4-х топовых моделей. Можно заметить следующее:
1. В двух моделей min_samples_leaf абсолютно идентичны (231), в то же время у других двух моделях значения схожи (44 и 27), но категорически отличаются от двух предыдущих
2. По критерию разбиения тоже есть различия: две модели используют энтропию, другие две - gini
3. Метрика, которая каждая модель улучшает, тоже разная, причем у всех
4. n_estimators у всех приблизительно одинаковый
5. У всех одинаковая глубина, она равна 10
6. Только одна модель использует bootstrap
7. min_samples_split лежит в диапазоне от 235 до 539

Из всего этого можно выбрать улучшаемую метрику roc_auc_score и следующие диапазоны гиперпараметров:
1. min_samples_leaf - значения меньше 100, в районе 40
2. criterion='entropy'
3. n_estimators - [190, 240]
4. max_features='sqrt'
5. max_depth - [9, 11]
6. bootstrap=False
7. min_samples_split - [417, 497]
